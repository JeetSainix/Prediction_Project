```python
df.drop('locality',axis=1,inplace=True)
df
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
categorical_cols=['type','price_unit','status','status','region','age']
for col in categorical_cols:
    df[col]=label_encoder.fit_transform(df[col])
```

```python
df
```

```python
x=df.drop('price',axis=1)
y=df['price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.fit_transform(x_test)
x_train_scaled
```

# Model Training

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train_scaled,y_train)
y_pred=lr.predict(x_test_scaled)
y_pred
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
mean_absolute_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

```python
from sklearn.linear_model import Ridge
r=Ridge()
r.fit(x_train_scaled,y_train)
y_pred=r.predict(x_test_scaled)
y_pred
```

```python
from sklearn.metrics import r2_score,mean_absolute_error

print(mean_absolute_error(y_test,y_pred))

print(r2_score(y_test,y_pred))
```

```python
from sklearn.linear_model import Lasso
ls=Lasso()
ls.fit(x_train_scaled,y_train)
y_pred=ls.predict(x_test_scaled)
y_pred
```

```python
from sklearn.metrics import r2_score,mean_absolute_error

print(mean_absolute_error(y_test,y_pred))

print(r2_score(y_test,y_pred))
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_absolute_error
xgb = XGBRegressor(
    n_estimators=800,
    learning_rate=0.03,
    max_depth=5,
    subsample=0.9,
    colsample_bytree=0.9
)
xgb.fit(x_train, y_train)
y_pred = xgb.predict(x_test)

# evaluate
print("XGB R2:", r2_score(y_test, y_pred))
print("XGB MAE:", mean_absolute_error(y_test, y_pred))
```

```python
import pickle
pickle.dump(xgb,open("D:/Model/house_price.pkl","wb"))
```
